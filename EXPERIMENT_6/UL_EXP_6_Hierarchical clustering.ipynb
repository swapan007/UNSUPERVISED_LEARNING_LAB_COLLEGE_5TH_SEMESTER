```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
datasubset = dataset.loc[:,["Age","Annual Income (k$)"]]
```

```python
plt.figure(figsize=(10,7))
plt.scatter(datasubset[["Age"]], datasubset[["Annual Income (k$)"]], s=100, c='blue')
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10,7))


dendrogram= sch.dendrogram(sch.linkage(datasubset,method="ward"))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('E.D')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean',linkage='ward')
cluster.fit_predict(datasubset)
```

```python
cl=cluster.fit_predict(datasubset)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(datasubset,cl)
```

```python
X = datasubset.values
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X[cl==0,0], X[cl==0,1], s=100, c='red',label='cluster1')
plt.scatter(X[cl==1,0], X[cl==1,1], s=100, c='blue',label='cluster2')
plt.title('cluster of mall customers')
plt.xlabel('age')
plt.ylabel('ANNUAL INCOME')
plt.show()
```

```python
from sklearn.cluster import KMeans
individual_clustering_score=[]
for i in range(1,15):
    kmeans= KMeans(n_clusters=i,init='random',random_state=42)
    kmeans.fit(X)
    individual_clustering_score.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(1,15),individual_clustering_score)
plt.title('Elbow method')
plt.xlabel('no.of clusters')
plt.ylabel('clustering score')
plt.show()
```

```python
kmeans=KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
pred= kmeans.predict(X)
pred
```

```python
silhouette_score(X,pred)
```
